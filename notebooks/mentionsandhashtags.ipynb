```python
mentions_list = []
network_list =[]
for each in tqdm(range(len(df.text))):
    source = df.user_screen_name[each]
    if pd.notnull(df.text[each]):
        input_string = df['text'][each]
        result = re.findall("@([a-zA-Z0-9]{1,15})", input_string)
        for i in range(len(result)):
            network_list.append(tuple((source, result[i], 'Mention')))
            mentions_list.append(tuple((source, result[i], 'Mention')))
    else:
        continue
```

```python
mentions_list
```

```python
hashtag_list = []
for each in tqdm(range(len(df.text))):
    source = df.user_screen_name[each]
    if pd.notnull(df.text[each]):
        input_string = df['text'][each]
        result = re.findall("#([a-zA-Z0-9]{1,15})", input_string)
        for i in range(len(result)):
            network_list.append(tuple((source, result[i], 'Hashtag')))
            hashtag_list.append(tuple((source, result[i], 'Hashtag')))
    else:
        continue
```

```python
len(hashtag_list) + len(mentions_list) == len(network_list)
```

```python
network_list = pd.DataFrame(network_list, columns = ['source', 'target', 'filter'])
```

```python
network_list.to_csv('network_list.csv')
```
